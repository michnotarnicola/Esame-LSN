### Scopo dell'esercizio

Si chiede di simulare un fluido di Lennard Jones nell'ensemble canonico:
1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ ($r_c = 2.2$)
2. fase liquida: $\rho^\star = 0.8$, $T^\star = 1.1$ ($r_c = 2.5$)
3. fase gassosa: $\rho^\star = 0.05$, $T^\star = 1.2$ ($r_c = 5.0$).

È richiesto di effettuare una stima compiuta sulla lunghezza dei blocchi della simulazione, calcolando dapprima la funzione di *autocorrelazione* e, in seguito, plottando le incertezze statistiche ottenute dalla blocking average in funzione della lunghezza dei blocchi. Inoltre, viene chiesto di inserire (sia in questo codice, sia in quello dell'esercitazione 4) il calcolo della *funzione di distribuzione a due corpi* $g(r)$.

### Discussione del lavoro

**Scrittura dei valori istantanei:** Nella funzione *Measure* si è aggiunta la scrittura sui file *V_inst.dat* e *P_inst.dat* dei valori istantanei di energia potenziale e pressione, dati necessari per l'equilibrazione e la stima della lunghezza dei blocchi.

**Calcolo della g(r):** Si è programmato il calcolo della funzione di distribuzione a coppie estendendo la size del *walker* di 100 componenti. Ciascuna componente rappresenta un bin di un'istogramma ( $N_{bins}=100$ ) che si estende sull'intervallo $r \in [0, L/2]$. Nella funzione *Measure* ogni qual volta che viene calcolata la distanza $dr$ tra due particelle, la si colloca nel bin a cui appartiene incrementando di 2 il valore di quella componente del walker. La funzione *Averages*, infine, calcola per ogni blocco la stima del numero di conteggi nei bin, le somma negli accumulatori *glob_av* (anche questi incrementati di 100 slot) e scrivendo su file i valori di ogni blocco, le medie cumulate e le incertezze su esse.

### <span style="color:red"> Simulazione 1: fase solida
    
La prima fase necessaria è quella dell'**equilibrazione**. Si è rilanciato più volte il programma con la procedura di ripartenza già introdotta nell'esercitazione 04 e, guardando i valori istantanei di energia potenziale e pressione, si è considerato il sistema equilibrato nel momento in cui tali grandezze mostravano oscillazioni intorno a un valore stabile. Per il solido, partendo da una configurazione FCC, si è raggiunto l'equilibrio con una sola simulazione di 1000 step. <p>
A questo punto, con la configurazione equilibrata, si è avviata una lunga simulazione di $M=2 \cdot 10^5$ passi. Con i valori istantanei ottenuti si è calcolata la **funzione di autocorrelazione**
$$ \chi(t)= \frac{ \frac{1}{t_M-t} \sum_{t'=0}^{t_M-t} X(t') X(t+t') - \bigl( \frac{1}{t_M-t} \sum_{t'=0}^{t_M-t} X(t') \bigr) \bigl( \frac{1}{t_M-t} \sum_{t'=0}^{t_M-t} X(t+t') \bigr)}{\frac{1}{t_M} \sum_{t'=0}^{t_M} X^2(t') - \bigl( \frac{1}{t_M} \sum_{t'=0}^{t_M-t} X(t') \bigr)^2}$$
per le grandezze $X =\{V/N, p\}$. Siccome il tempo macchina richiesto è estremamente lungo, si riportano le linee di codice che programmano tali operazioni, mentre il grafico finale viene caricato in foto.

```python
## NB: non compilare perchè richiede moltissimo tempo per produrre l'output
%matplotlib inline
from numpy import *
from math import *
import pylab as plt

def autocorr (x,t,tm):
    A,B,C,D,E = 0,0,0,0,0
    for s in range(tm-t):
        A+=x[s]*x[s+t]
        B+=x[s]
        C+=x[s+t]
    A/=float((tm-t))
    B/=float((tm-t))
    C/=float((tm-t))
    for s in range(tm):
        D+=x[s]**2
        E+=x[s]
    D/=float(tm)
    E/= float(tm)
    return (A-B*C)/(D-E*E)

V = genfromtxt('V_inst.dat')
P = genfromtxt('P_inst.dat')

time=arange(0,200,1)
chiV=zeros(200)
chiP=zeros(200)
for t in range(200):
    chiV[t]=autocorr(V,t, 200000)
    chiP[t]=autocorr(V,t, 200000)

plt.figure(figsize=(8.,6.))
plt.plot(time, chiV, label='Potential Energy')
plt.plot(time, chiP, label='Pressure')
plt.title('Autocorrelation functions: SOLID')
plt.xlabel('# MC step')
plt.ylabel('$\chi$(t)')
plt.legend()
plt.grid(True)
plt.savefig('ChiSOLID.png')
plt.show()
```

La funzione parte da 1 e decade verso 0 dopo un numero di step Monte Carlo tra i 50 e 100. I risultati delle due grandezze sono perfettamente sovrapponibili.
Questo suggerisce che la lunghezza dei blocchi debba essere molto più grande di tale valore. <p>
Per avere una idea più concreta di quanto debba essere lungo un blocco, si è implementato il codice sottostante che, al variare di $L \in [10, 5 \cdot 10^3]$ calcola le incertezze statistiche su $V$ e $p$.

```python
%matplotlib inline
from numpy import *
from math import *
import pylab as plt

def blockingaverage(X,M,L):
    N=int(M/L)
    ave = zeros(N)
    av2 = zeros(N)
    stima = 0
    stima2 = 0
    err = 0

    for i in range(N):
        somma = 0.0
        for j in range(L):
            k = j+i*L
            ave[i] += X[k]
        ave[i] /= L       # r_i 
        av2[i] = (ave[i])**2 # (r_i)^2 

    for i in range(N):
        stima += ave[i] # SUM_{j=0,i} r_j
        stima2+= av2[i] # SUM_{j=0,i} (r_j)^2
    stima/=N # Cumulative average
    stima2/=N # Cumulative square average
    err = sqrt((stima2 - stima**2)/N) # Statistical uncertainty
    return stima , err

V = genfromtxt('SOLID_V_inst.dat')
P = genfromtxt('SOLID_P_inst.dat')  

L=arange(10,5000, 50) #lenght of a block
V_medio=zeros(size(L))
p_medio=zeros(size(L))
sigma_V=zeros(size(L))
sigma_p=zeros(size(L))

for i in range(size(L)):
    V_medio[i] , sigma_V[i] = blockingaverage(V,size(V),L[i])
    p_medio[i] , sigma_p[i] = blockingaverage(P,size(P),L[i])

fig = plt.figure(figsize=(12, 3.8))

ax = fig.add_subplot(121)
ax.plot(L, sigma_V, label='$\sigma$_V')
ax.plot(L, sigma_p, label='$\sigma$_p')
ax.set_title('Statistical uncertainties: SOLID')
ax.set_xlabel('Lenght of the block L')
ax.grid(True)
ax.legend()

ax = fig.add_subplot(122)
ax.set_xlim(0,1000)
ax.plot(L, sigma_V, label='$\sigma$_V')
ax.plot(L, sigma_p, label='$\sigma$_p')
ax.set_title('Statistical uncertainties: SOLID')
ax.set_xlabel('Lenght of the block L')
ax.grid(True)
ax.legend()

plt.show()
```

Le due figure raffigurano la stessa grandezza, ma la seconda evidenzia solo i dati fino a $L=1000$. Quello che si osserva è che in una prima fase le incertezze tendono a crescere. Questo perchè i blocchi sono troppo corti e le incertezze ottenute in ipotesi di CLT sono sottostimate. All'aumentare del numero di blocchi, invece, il valore delle incertezze fluttua, ma si stabilizza attorno a un valore costante. La saturazione avviene a parire da $L=500$. Questo indica che la lunghezza dei blocchi è tale da scorrelarli tra loro e le previsioni del CLT diventano corrette. In questa regione è ininfluente il particolare valore di $L$ perchè le incertezze saranno sempre simili tra loro. <p>
Pertanto, da questa analisi si determinano i parametri della simulazione vera e propria: $N=30$ e $L=500$. I risultati ottenuti per la fase solida sono riportati qui sotto.

```python
%matplotlib inline
from numpy import *
from math import *
import pylab as plt
from scipy.optimize import curve_fit

def fitfunc(x,c):
    return c

gofr= genfromtxt('SOLID_output.gave.0')
gdir= genfromtxt('SOLID_ave_gofr.dat')
V= genfromtxt('SOLID_output.epot.0')
p= genfromtxt('SOLID_output.pres.0')

eps=120*1.38 # per 10^-23 Joule
sigma =0.34 #nanometres
press= 4.213 # per 10^7 Pa
Temp= 120 #Kelvin

plt.errorbar(V[:,0], eps*V[:,2], yerr=eps*V[:,3], label='NVT')
p_opt, p_cov = curve_fit(fitfunc, V[15:,0], eps*V[15:,2], bounds=(-1252,-1250))
plt.title('Potential energy : SOLID')
plt.xlabel('# blocks')
plt.ylabel('V (x 10^-23 J)')
plt.grid(True)
plt.legend()
plt.show()
print('Fit del valore di convergenza      V=',p_opt[0], '\pm', sqrt(p_cov[0]), 'x 10^-23 J')

plt.errorbar(p[:,0], press*p[:,2], yerr=press*p[:,3], label='NVT')
p_opt, p_cov = curve_fit(fitfunc, p[15:,0], press*p[15:,2], bounds=(31.6,32))
plt.title('Pressure : SOLID')
plt.xlabel('# blocks')
plt.ylabel('p (x 10^7 Pa)')
plt.grid(True)
plt.legend()
plt.show()
print('Fit del valore di convergenza      p=',p_opt[0], '\pm', sqrt(p_cov[0]), 'x 10^7 Pa')

plt.errorbar(sigma*gofr[:,0], gofr[:,1], yerr=gofr[:,2], label='NVT')
plt.errorbar(sigma*gdir[:,0], gdir[:,1], yerr=gdir[:,2], label='NVE')
plt.title('Radial couple distribution function g(r) : SOLID')
plt.xlabel('r (nm)')
plt.ylabel('g(r)')
plt.grid(True)
plt.legend()
plt.show()
```

Nelle prime due immagini sono riportate le stime di energia potenziale e pressione, con relativo fit dei valori di trend. I risultati concordano tra loro:

    Questi picchi sono di ampiezza variabile e all'aumentare della distanza si riducono di intensità e tendono a oscillare intorno a 1.

### <span style="color:red"> Simulazione 3: fase gassosa

Per la fase gassosa, la prima operazione da fare è scegliere il passo *Delta* del Metropolis per garantire un'accettazione del $50\%$. Si è operato per tentativi, osservando che i parametri utilizzati per le fasi solida e liquida, pari rispettivamente a $0.2$ e $1.2$ se applicati alla fase gassosa (molto meno densa) comportavano un'accettazione superiore al $95 \%$. La prima conclusione è che il parametro *Delta* per il gas dovrebbe essere significativamente maggiore di 1. D'altro canto, l'algoritmo lavora in PBC, per cui la distanza tra il punto nuovo proposto $\textbf{r}_{new}$ e il punto vecchio $\textbf{r}_{old}$ non dovrebbe eccedere $L_{box}/2$. Siccome la velocità è un numero random tra -0.5 e 0.5, implica una scelta per *Delta* pari a $ L_{box}$. Questo comporta un'accettazione di circa il $61 \%$, che rappresenta il rate più basso possibile con questi parametri.

L'**equilibrazione** è stata svolta a partire dalla configurazione liquida ed è constata di 6 rilanci da 1000 step l'uno. La funzione di autocorrelazione decade a 0 dopo 20 step, molto meno che negli altri due casi. Questo è sensato, perchè rispetto alle altre due fasi il parametro *Delta* scelto è significativamente più grande e il sistema compie ampie mosse, per cui si scorrela molto velocemente. L'analisi dell'andamento delle sigma, invece, mostra saturazione per $L=100$.
