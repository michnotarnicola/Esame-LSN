## <span style="color:blue"> Esercizio 1 & 2 
(code VariationalMC.cpp)


### Scopo dell'esercizio

Si richiede di scrivere un codice di Monte Carlo variazionale per simulare una particella quantistica in modo unidimensionale in un potenziale $V(x) = x^4 - \frac{5}{2}x^2$. <p>
Scegliendo come funzione di prova $\Psi_T(x) = e^{-\frac{(x-\mu)^2}{2\sigma^2}}+e^{-\frac{(x+\mu)^2}{2\sigma^2}}$, si chiede di calcolare $$\langle H \rangle_{\mu \sigma} = \frac{\langle \Psi_T | H | \Psi_T\rangle}{\langle \Psi_T | \Psi_T \rangle}$$ e di implementare un algoritmo per cercare i valori di $\mu$ e $\sigma$ che minimizzano tale valore di aspettazione. <p>
 Il risultato ottenuto sarà la miglior approssimazione del ground state per fissata $\Psi_T$.

### Descrizione del codice

Il programma opera la ricerca del valore di aspettazione minimo dell'hamiltoniana utilizzando una tecnica di **simulated annealing**. <p>
Si sono fissati dei parametri iniziali $\mu=1$ e $\sigma=1$ e si associato al sistema una temperatura fittizia pari a $T_i=0.1$. Questo valore è stato scelto dopo una serie di rilanci del programma ed è ragionevole perchè la densità di probabilità iniziale $\Psi_T(\mu=1, \sigma=1)$ assume il valore massimo in un intervallo compreso tra i due minimi del potenziale. In altre parole, lo stato iniziale è in una regione dello spazio dei parametri $\mu$ e $\sigma$ alquanto favorevole per la ricerca del minimo dell'energia e non è necessario impostare una temperatura troppo elevata.

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

def Vpot(x):
    return (x**2 - 2.5)*x**2

mu = 1
sigma = 1
a = 10
N = 1000 # number of iterations

x = linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

prob=zeros(N)
for i in range(N):
    prob[i]= ( exp(-(x[i]-mu)**2/(2*sigma**2)) + exp(-(x[i]+mu)**2/(2*sigma**2)) )**2
norm= sum(prob)*dx
         
plt.figure()
scale = 0.3
plt.xlim(-2.5,2.5)
plt.ylim(-0.5,0.5)
plt.plot(x, scale*V, color="Black", label='V(x)')
plt.plot(x, prob/norm, label ='|$\psi$|^2')
plt.grid(True)
plt.legend()
plt.show()
```

A partire dalla temperatura iniziale, si è raffreddato il sistema sino a $T_f=0.01$, operazione svolta in $N_{step}=50$ step totali.
In ogni step $i_{step}$ la temperatura è stata posta pari a $T = T_i- \frac{T_i-T_f}{N_{step}} \cdot i_{step}$ e si è proceduto nel seguente modo. <p>
Si è lasciato equilibrare il sistema alla temperatura $T$ e, in seguito, si è calcolato $\langle H \rangle_{\mu, \sigma}$. Questo è stato svolto dalla funzione *Move* implementando lo stesso algoritmo di Metropolis dell'esercitazione 5 con i parametri $N_{blk}=50$, $L_{blk}=2000$, $N_{eq}=1000$, $N_{step \space per \space blocco}=50$. Il valore di $\langle H \rangle_{\mu, \sigma}$ corrisponde al valore finale della media di blocco cumulativa. <p>
Dopo di che, si è proposto un nuovo valore per i parametri variazionali $\mu'$ e $\sigma'$, scelto casualmente con distribuzione uniforme centrata su $\mu$ e $\sigma$, si è calcolato $\langle H \rangle_{\mu', \sigma'}$ identicamente a prima e si è accettata la nuova configurazione con probabilità $p=\min[ 1, e^{-\beta \space (\langle H \rangle_{\mu', \sigma'}-\langle H \rangle_{\mu, \sigma})}]$. Questa procedura è stata ripetuta per $N_{cycle}=50$ volte. <p>

Abbassando molto lentamente la temperatura, in modo che il sistema non perda ergodicità, si vede che le mosse accettate saranno sempre più frequentemente quelle che riducono il valor medio di H, fino ad arrivare al valore minimo. <p>
Una volta ottenuti i valori ottimali di $\mu$ e $\sigma$ si è avviata la simulazione vera e propria (*Simulation*), re-implementando il Metropolis e scrivendo su file i valori di punti campionati e l'andamento delle medie cumulate.

### Analisi dati
In basso è riportato l'andamento del $\langle H \rangle$ in funzione del numero di step $i_{step}$ di run della procedura di annealing.

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

Anneal = genfromtxt('AndamentoH.dat')
plt.figure()
plt.plot(Anneal[:,0], Anneal[:,1])
plt.xlabel('# step di temperatura')
plt.ylabel('<H>')
plt.grid(True)
plt.show()
```

Si osserva un andamento molto frastagliato dovuto al *rumore termico* prodotto dall'algoritmo, ma negli ultimi step si inizia a osservare una saturazione a un valore costante, segnale che il sistema si sta muovendo attorno al minimo.<p>
I valori finali ottenuti per i parametri sono stati
$$ \mu= 0.786399 \qquad \sigma= 0.612139 .$$
Si è, dunque, graficata la funzione d'onda variazionale ottimizzata da tali parametri.

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

#------NUMERICAL DIAGONALIZATION OF H-----#
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = diag(ones(N-1),-1)-2*diag(ones(N),0)+diag(ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = linalg.eigh(H)

# Take the transpose & normalize
psi = transpose(psi)
psi = psi/sqrt(dx)

print("Ground state energy: ", E[0])

#---------------------------------#

# Plot histogram of the sampled data vs the theoretical probability density 
mu= 0.786399
sigma= 0.612139

densityprob = genfromtxt('Psi_T.dat')
n_bins = 30

prob=zeros(N)
for i in range(N):
    prob[i]= ( exp(-(x[i]-mu)**2/(2*sigma**2)) + exp(-(x[i]+mu)**2/(2*sigma**2)) )**2
norm= sum(prob)*dx

plt.figure(figsize=(8,5))
plt.hist(densityprob[:], n_bins, density=True, range=(-2,2), color='y')
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2, label="Numerical |$\psi$|^2", color='b')
plt.plot(x,prob/norm, label="Variational |$\psi$|^2", color='r')

plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.legend()
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```

In figura è mostrato l'istogramma degli x campionati con il **metodo variazionale** in raffronto ai risultati ottenuti da una **diagonalizzazione numerica** dell'hamiltoniana. Inoltre si riporta anche il valore dell'energia di stato fondamentale ottenuta da tale risoluzione numerica del problema. <p>
Quello che si osserva è che la funzione $\Psi_T$, ancorchè ottimizzata, non coincide con l'autofunzione "esatta", in particolare ha una larghezza inferiore e la probabilità intorno ai valori massimi è più alta. Ma questo è assolutamente aspettato perchè il metodo non fornisce il ground state esatto, bensì la sua miglior stima compatibilimente con la funzione di prova scelta. Ciò è evidente andando a guardare il valor medio dell'hamiltoniana ottenuto sullo stato $|\Psi_T\rangle$. 

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

E = genfromtxt('Hmedio.dat')
plt.figure()
plt.errorbar(E[:,0], E[:,2], yerr=E[:,3])
plt.xlabel('# blocchi')
plt.ylabel('<H>')
plt.grid(True)
plt.show()
```

Si osserva che $\langle H \rangle \simeq-0.445$, a fronte di un ground state esatto $E_{gs}\simeq -0.46$. Chiaramente $\langle H \rangle > E_{gs}$, come aspettato. Va notata, però la potenza del metodo variazionale, che arriva a deviazioni dal valore esatto $$\frac{\langle H \rangle-E_{gs}}{E_{gs}} \simeq 0.03 = 3\% $$

### Analisi dati
### <span style="color:red"> PIGS : simulazione a T=0

La simulazione a T=0 effettua un algoritmo proiettivo che, partendo da una funzione di prova $|\psi\rangle$ con overlap non nullo con il ground state $|GS\rangle$, ottiene un' approssimazione dello stato fondamentale tramite l'evoluzione per tempi immaginari.
Le due funzioni di prova sono molto diverse tra loro: la prima, quella costante, è significativamente diversa dal ground state, mentre la seconda gli è molto più simile. Questo implica che, perchè il PIGS raggiunga il ground state sarà necessaria un'evoluzione immaginaria di $\psi_1$ per tempi maggiori rispetto a $\psi_2$. Questo è stato riscontrato nella pratica:
- nel caso $\psi_1$ è stato necessario porre $\tau=8$ per ottenere una densità di probabilità che ben rappresenta lo stato fondamentale;
- nel caso $\psi_2$, invece, è stato sufficiente $\tau=3$, meno della metà del caso precedente. <p>

Il risultato finale, però, è in ambo i casi sovrapponibile perfettamente al ground state esatto. Questo mostra la potenza dell'algoritmo proiettivo e la sua maggiore efficacia rispetto a un semplice metodo variazionale.    

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

#------NUMERICAL DIAGONALIZATION OF H-----#
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = diag(ones(N-1),-1)-2*diag(ones(N),0)+diag(ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = linalg.eigh(H)

# Take the transpose & normalize
psi = transpose(psi)
psi = psi/sqrt(dx)

#---------------------------------#

# Plot histogram of the PIGS algorithm vs the numerical results

psi1=genfromtxt('probability.const')
psi2=genfromtxt('probability.trial')

plt.figure(figsize=(8,5))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2, label="Numerical |$\psi$|^2", color='b')
plt.plot(psi1[:,0],psi1[:,1], label="PIGS with $\psi_1$", color='r')
plt.plot(psi2[:,0],psi2[:,1], label="PIGS with $\psi_2$", color='g')

plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.legend()
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```

È bene sottolineare che l'istogramma delle densità di probabilità è stato svolto soltanto con i valori ottenuti tra la timeslice 120 e la 180.
La ragione di ciò è evidente guardando l'andamento di alcuni valori medi al variare della funzione di prova. Nel grafico sottostante sono riportati $\langle T \rangle$ e $\langle V \rangle$ in funzione del numero di step temporali dell'evoluzione. 

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

T1=genfromtxt('kinetic.const')
V1=genfromtxt('potential.const')
T2=genfromtxt('kinetic.trial')
V2=genfromtxt('potential.trial')

fig = plt.figure(figsize=(12, 3.8))

ax = fig.add_subplot(121)
ax.errorbar(T1[:,0],T1[:,1], yerr=T1[:,2], color='C0', label='$\psi_1$')
ax.errorbar(T2[:,0],T2[:,1], yerr=T2[:,2], color='C1', label='$\psi_2$')
ax.set_title('Kinetic energy')
ax.set_xlabel('# step')
ax.set_ylabel('<T>')
ax.legend()
ax.grid(True)

ax = fig.add_subplot(122)
ax.errorbar(V1[:,0],V1[:,1], yerr=V1[:,2], color='C0', label='$\psi_1$')
ax.errorbar(V2[:,0],V2[:,1], yerr=V2[:,2], color='C1', label='$\psi_2$')
ax.set_title('Potential energy')
ax.set_xlabel('# step')
ax.set_ylabel('<V>')
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```

L'andamento di $\langle V \rangle$ esemplifica il principio che sta alla base dell'algoritmo. Il valor medio è ottenuto facendo l'evoluzione immaginaria del bra e del ket, affinchè questi convergano verso il ground state del sistema. Fintanto che si è nelle prime e nelle ultime timeslices l'evoluzione immaginaria non è ancora arrivata a convergenza (rispettivamente per il ket e per il bra) e le previsioni per il valor medio di V sono molto distinte da quelle effettive. È soltanto nelle timeslices centrali che l'algoritmo arriva a previsioni confrontabili con l'andamento effettivo del sistema fisico in studio ed è per questa ragione che gli istogrammi delle densità di probabilità sono stati ristretti nell'intervallo $[120,180]$. L'effetto è molto più evidente per la funzione $\psi_1$ rispetto alla $\psi_2$, perchè la funzione di prova scelta è significatvamente diversa dal ground state vero e questo richiede tempi di evoluzione maggiori per "ripulire" la funzione dalla sovrapposizione con gli altri autostati dell'hamiltoniana. L'andamento per la funzione $\psi_2$ è, invece, sostanzialmente piatto perchè la funzione ha un overlap maggiore con il ground state. Ad ogni modo, nell'intervallo $[120,180]$ si osserva che le previsioni ottenute nei due casi sono statisticamente equivalenti. <p>

È bene sottolineare anche che, nonostante il tempo $\tau$ sia stato diverso nei due casi, il numero di timeslices è stato mantenuto per comodità sempre pari a $300$. Chiaramente se un passo discreto $\Delta \tau=8/300$ è ritenuto sufficiente per applicare l'approssimazione primitiva, lo sarà anche $\Delta \tau=3/300$, ma questa ulteriore riduzione del timestep potrebbe indurre delle fluttuazioni ulteriori sui risultati della simulazione con $\psi_2$. In ogni caso, visto che dal nostro punto di vista è sufficiente un'analisi piuttosto qualitativa dei grafici, ciò non risulta un grave problema.

### <span style="color:red"> PIMC : simulazione a T>0

Nel caso di temperatura finita lo stato del sistema sarà uno stato termico, ovvero uno stato miscela $\rho= e^{-\beta H}/Z$. In termini di sviluppo spettrale sugli autostati di H, $\rho= \frac{1}{Z}\sum_{n=0}^{\infty} e^{-\beta E_n} |n\rangle\langle n| \simeq \frac{1}{Z} \biggl[ e^{-\beta E_0} |0\rangle\langle 0| + e^{-\beta E_1} |1\rangle\langle 1| + ... \biggr]$. <p>
Questo significa che lo stato è una combinazione lineare di stati puri e, dunque, la densità di probabilità cambia al variare della temperatura. Per osservare come viene modificata la densità di probabilità, è stato simulato il sistema a tre temperature via via crescenti: $T=1,10,100$, confrontando gli istogrammi ottenuti con quelli a temperatura nulla. 

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

p1T0=genfromtxt('probability.const')
p1T1=genfromtxt('p_T1.const')
p1T10=genfromtxt('p_T10.const')
p1T50=genfromtxt('p_T100.const')
p2T0=genfromtxt('probability.trial')
p2T1=genfromtxt('p_T1.trial')
p2T10=genfromtxt('p_T10.trial')
p2T50=genfromtxt('p_T100.trial')

fig = plt.figure(figsize=(12, 3.8))

ax = fig.add_subplot(121)
ax.plot(p1T0[:,0],p1T0[:,1], label= 'T=0')
ax.plot(p1T1[:,0],p1T1[:,1], label= 'T=1')
ax.plot(p1T10[:,0],p1T10[:,1], label= 'T=10')
ax.plot(p1T50[:,0],p1T50[:,1], label= 'T=100')
ax.set_title('Densità di probabilità. Caso 1: Psi trial = $\psi_1$')
ax.set_xlabel('x')
plt.legend()
ax.grid(True)

ax = fig.add_subplot(122)
ax.plot(p2T0[:,0],p2T0[:,1], label= 'T=0')
ax.plot(p2T1[:,0],p2T1[:,1], label= 'T=1')
ax.plot(p2T10[:,0],p2T10[:,1], label= 'T=10')
ax.plot(p2T50[:,0],p2T50[:,1], label= 'T=100')
ax.set_title('Densità di probabilità. Caso 2: Psi trial = $\psi_2$')
ax.set_xlabel('x')
plt.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```

Quello che si osserva è che la distribuzione di probabilità di ground state, piccata sulle posizioni x minime, al crescere di T tende ad abbassarsi e a diventare più larga. Questo perchè la particella quantistica a temperatura finita è in una sovrapposizione di autostati dell'hamiltoniana. La cosa si intuisce osservando la struttura dei primi stati eccitati (nella figura in basso). <p>
Il primo stato eccitato ha un nodo nell'origine e questo spiega perchè a T=1 la densità di probabilità intorno all'orgine è inferiore al caso del puro ground state. Si può affermare che per T=1 il sistema sia sostanzialmente in una sovrapposizione degli stati $|0\rangle$ e $|1\rangle$. Le probabilità di occupazione dei livelli più alti sono trascurabili. <p>
La situazione cambia notevolmente per $T=10$ e $T=100$, dove il contributo del terzo e del quarto livello energetico diventano visibili, con il risultato che la distribuzione di probabilità tende ad appiattirsi sempre di più. La sovrapposizione di tante funzioni oscillanti pian piano tende a una distribuzione quasi piatta.<p>
All'aumentare ulteriore della temperatura ci si aspetta che diventino rilevanti anche le probabiità di occupazione dei livelli successivi, con il risultato di una densità di probabilità sempre più distinta da quella di ground state.

```python
%matplotlib inline
from numpy import *
from math import *
import matplotlib
import matplotlib.pyplot as plt

#------NUMERICAL DIAGONALIZATION OF H-----#
def Vpot(x):
    return (x**2 - 2.5)*x**2
    #return 0.5*x**2

hbar = 1
m = 1
a = 10
N = 1000 # number of iterations

# Step sizes
x = linspace(-a/2, a/2, N)
dx = x[1] - x[0] # the step size
V = Vpot(x)

# The central differences method: f" = (f_1 - 2*f_0 + f_-1)/dx^2

CDiff = diag(ones(N-1),-1)-2*diag(ones(N),0)+diag(ones(N-1),1)
# np.diag(np.array,k) construct a "diagonal" matrix using the np.array
# The default is k=0. Use k>0 for diagonals above the main diagonal, 
# and k<0 for diagonals below the main diagonal

# Hamiltonian matrix
H = (-(hbar**2)*CDiff)/(2*m*dx**2) + diag(V)

# Compute eigenvectors and their eigenvalues
E,psi = linalg.eigh(H)

# Take the transpose & normalize
psi = transpose(psi)
psi = psi/sqrt(dx)
print("Ground state energy: ", E[0])
print("1st excited state energy: ", E[1])
print("2nd excited state energy: ", E[2])
print("3rd excited state energy: ", E[3])

plt.figure(figsize=(8,5))
scale = 0.3
plt.plot(x, scale*V, color="Black", label="Potential") # plot the potential
plt.plot(x,(psi[0])**2, label="|$\psi_0$|^2")
plt.plot(x,(psi[1])**2,label="|$\psi_1$|^2")
plt.plot(x,(psi[2])**2, label="|$\psi_2$|^2")
plt.plot(x,(psi[3])**2, label="|$\psi_3$|^2")

plt.title("Potential & Probabilities")
plt.xlabel("x")
plt.grid(True)
plt.legend()
plt.xlim((-3,3))
plt.ylim((-0.6,0.6))
plt.show()
```
