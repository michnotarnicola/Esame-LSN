### <span style="color:red"> Preminaries: upload delle librerie e del dataset MNIST

```python
import tensorflow as tf
from tensorflow import keras
import os
import numpy as np
import matplotlib.pyplot as plt
seed=0
np.random.seed(seed) # fix random seed
tf.random.set_seed(seed)
```

```python
from tensorflow.keras.datasets import mnist

# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 
# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()

print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

## <span style="color:blue"> Esercizio 1

### Scopo dell'esercizio

Lo scopo dell'esercizio è quello di testare la rete DNN presentata per riconoscimento delle cifre da $0$ a $9$, analizzandone il funzionamento al variare dei possibili ottimizzatori. <p>
    
### Discussione del lavoro   
### <span style="color:red"> Fase 1: definizione della rete neurale profonda DNN
Qui di seguito sono riportate le linee di codice che creano e inizializzano la rete feed forward utilizzata. <p>
L'input corrisponde a un'immagine 28x28 "srotolata" in un array unimensionale di size 784, mentre l'output consiste di un intero tra 0 e 9, vale a dire la previsione della cifra rappresentata nel dato in ingresso. <p>
La rete ha un'**architettura** tipo $784 \rightarrow 400 \rightarrow 100 \rightarrow 100 \rightarrow 10$. Gli hidden layer hanno come funzione di attivazione la *ReLu*, mentre il layer di output usa una *Softmax*, visto che ha da risolvere un problema di classificazione. <p>
La funzione loss scelta è stata la *Categorical Cross Entropy*, come metrica invece la *Accuracy*. L'ottimizzatore è stato variato esplorando varie possibilità. Qui di seguito è stato mantenuto lo *Stochastic Gradient Discent*.

```python
# reshape data, it could depend on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows*img_cols)
X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)

# cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

##------ look at an example of data point------#
#print('an example of a data point with label', Y_train[20])
# matshow: display a matrix in a new figure window
#plt.matshow(X_train[20,:].reshape(28,28),cmap='binary')
#plt.show()

##------print('an example of a data point with label', Y_train[20], 'before to_categorical ...')------#
# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
#print('... and with label', Y_train[20], 'after to_categorical')
#print()
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

def create_DNN():
    # instantiate model
    model = Sequential()
    # add a dense all-to-all relu layer
    model.add(Dense(400,input_shape=(img_rows*img_cols,), activation='relu'))
    # add a dense all-to-all relu layer
    model.add(Dense(100, activation='relu'))
    # apply dropout with rate 0.5
    model.add(Dropout(0.5))
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    return model

print('Model architecture created successfully!')
```

```python
from tensorflow.keras.optimizers import SGD, Adam, RMSprop, Adagrad, Adadelta, Adam, Adamax, Nadam

def compile_model():
    # create the model
    model=create_DNN()
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer=SGD(),
                  metrics=['accuracy'])
    return model

# create the deep neural net
model_DNN = compile_model()

print('Model compiled successfully and ready to be trained.')
```

### <span style="color:red"> Fase 2: training del modello

Una volta creata la rete, si è proceduto alla fase di training. Il dataset di training è stato diviso in minibatches di 32 dati e il numero di epoche scelto è stato pari a 20.

```python
# training parameters
batch_size = 32
epochs = 20

# train DNN and store training info in history
history = model_DNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

### <span style="color:red"> Fase 3: applicazione della rete 'trainata' al dataset di test

Di seguito sono riportati l'andamento della funzione loss e accuracy dei due dataset in funzione del numero di epoche.

```python
# evaluate model
score = model_DNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Optimizer: SGD')
plt.ylabel('Model accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.grid()
#plt.savefig('accuracySGD.png')
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Optimizer: SGD')
plt.ylabel('Model loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.grid()
#plt.savefig('lossSGD.png')
plt.show()
```

La funzione costo sul dataset di training tende a decrescere con l'aumentare delle epoche e ad avvicinarsi alla curva del dataset di test. Questo è il segnale che la rete sta imparando, ma non è ancora giunta alla fase di overfitting. I dati di test mostrano una loss inferiore solamente perchè per loro non si attiva il layer di dropout. Lo stesso ragionamento a rovescio si può fare per il grafico della accuracy. <p>
Infine, in basso sono riportati i risultati predetti dalla rete su alcuni dati del dataset di test. Come si piò osservare, tutte le lettere sono riconosciute, segnale di un buon apprendimento.

```python
#X_test = X_test.reshape(X_test.shape[0], img_rows*img_cols)
predictions = model_DNN.predict(X_test)

X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols,1)

plt.figure(figsize=(15, 15)) 
for i in range(10):    
    ax = plt.subplot(2, 10, i + 1)    
    plt.imshow(X_test[i, :, :, 0], cmap='gray')    
    plt.title("Digit: {}\nPredicted:    {}".format(np.argmax(Y_test[i]), np.argmax(predictions[i])))    
    plt.axis('off') 
plt.show()
```

## <span style="color:blue"> Esercizio 2 & 3

### Scopo dell'esercizio

Lo scopo dell'esercizio è quello di risolvere il problema delle cifre implementando una rete CNN e, dopo averla trainata, di testarla su cifre scritte a mano da noi stessi. <p>
    
### Discussione del lavoro   
### <span style="color:red"> Fase 1: definizione della rete neurale convoluzionale CNN
Si è ripartiti dal dataset caricato da MNIST e, questa volta, gli input sono stati riscritti sottoforma di tensori 3D 28x28x1(dove 1 rappresenta il fatto che si lavora in scala di grigi). Gli output, chiaramente, sono ancora interi da 0 a 9. L'architettura della rete è stata ispirata da quella del sitoweb www.cs.ryerson.ca~aharley/vis/conv/flat.html, del tipo $1 \rightarrow 10 \rightarrow 10 \rightarrow 16 \rightarrow 16\rightarrow 400 \rightarrow 100 \rightarrow 100 \rightarrow 10$. Lo schema della rete è riportato in basso. Come loss e accuracy sono state mantenute quelle precedenti e come ottimizzatore si è ancora scelto lo SGD.

```python
# you will need the following for Convolutional Neural Networks
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D

# reshape data, depending on Keras backend
if keras.backend.image_data_format() == 'channels_first':
    X_train = X_train.reshape(X_train.shape[0], 1, img_rows, img_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, img_rows, img_cols)
    input_shape = (1, img_rows, img_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
    input_shape = (img_rows, img_cols, 1)

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
    
print('X_train shape:', X_train.shape)
print('Y_train shape:', Y_train.shape)
print()
print(X_train.shape[0], 'train samples')
print(X_test.shape[0], 'test samples')
```

```python
def create_CNN():
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 6 filtersv ----- 8 16 120 dropout
    #model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
    #model.add(MaxPooling2D(pool_size=(2, 2)))
    #model.add(Conv2D(32, kernel_size=(3,3), activation='relu'))
    #model.add(MaxPooling2D(pool_size=(2, 2)))
    #model.add(Dropout(0.5))
    #model.add(Flatten())
    #model.add(Dense(128, activation='relu'))
    #model.add(Dropout(0.5))
    #model.add(Dense(num_classes, activation='softmax'))
    model.add(Conv2D(8, kernel_size=(3, 3), activation='relu', input_shape=input_shape))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    model.add(Conv2D(16, kernel_size=(3,3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2, 2)))
    #model.add(Dropout(0.5))
    model.add(Flatten())
    model.add(Dense(120, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(num_classes, activation='softmax'))


    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD',
                  metrics=['accuracy'])
    return model

# create the deep conv net
model_CNN=create_CNN()

print('Model compiled successfully and ready to be trained.')
```

```python
model_CNN.summary()
```

### <span style="color:red"> Fase 2: training del modello

Il dataset di training è stato diviso ancora in minibatches di 32 dati e il numero di epoche scelto è stato pari a 20.

```python
# training parameters
batch_size = 5
epochs = 7

# train CNN
history = model_CNN.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs, shuffle=True,
          verbose=1,
          validation_data=(X_test, Y_test))
```

### <span style="color:red"> Fase 3: applicazione della rete 'trainata' al dataset di test e alle cifre scritte a mano

Di seguito sono riportati l'andamento della funzione loss e accuracy dei due dataset in funzione del numero di epoche.

```python
# evaluate model
score = model_CNN.evaluate(X_test, Y_test, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history

# summarize history for accuracy
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Optimizer: SGD')
plt.ylabel('Model accuracy')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.grid()
plt.show()

# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Optimizer: SGD')
plt.ylabel('Model loss')
plt.xlabel('Epoch')
plt.legend(['train', 'test'], loc='best')
plt.grid()
plt.show()
```

Si conferma l'andamento decrescente della loss e crescente della accuracy, con le due curve di test e train che tendono ad avvicinarsi tra loro. <p>
    
Come azione finale, la rete è stata applicata sia ai dati di test, sia alle lettere scritte a mano dal sottoscritto. Il risultato sui 10 **dati estratti dal dataset di MNIST** è il seguente:

Display your filters:

```python
# layer_index should be the index of a convolutional layer
layer_index=0
# retrieve weights from the convolutional hidden layer
filters, biases = model_CNN.layers[layer_index].get_weights()
# normalize filter values to 0-1 so we can visualize them
f_min, f_max = filters.min(), filters.max()
filters = (filters - f_min) / (f_max - f_min)
print(filters.shape)

# plot filters
n_filters, ix = filters.shape[3], 1
for i in range(n_filters):
    # get the filter
    f = filters[:, :, :, i]
    # specify subplot and turn of axis
    ax = plt.subplot(1,n_filters, ix)
    ax.set_xticks([])
    ax.set_yticks([])
    # plot filter channel in grayscale
    plt.imshow(f[:, :, 0], cmap='gray')
    ix += 1
# show the figure
plt.show()
```

### <span style="color:green">Supplementary information 3: Monitor layer outputs in your CNN</span>

```python
test_index = 0

from tensorflow.keras.models import Model
layer_outputs = [layer.output for layer in model_CNN.layers]
activation_model = Model(inputs=model_CNN.input, outputs=layer_outputs)
activations = activation_model.predict(X_test[test_index].reshape(1,28,28,1))
 
def display_activation(activations, col_size, row_size, layer_index): 
    activation = activations[layer_index]
    activation_index=0
    fig, ax = plt.subplots(row_size, col_size, figsize=(row_size*3,col_size*3))
    for row in range(0,row_size):
        for col in range(0,col_size):
            ax[row][col].imshow(activation[0, :, :, activation_index], cmap='gray')
            activation_index += 1
```
